```python
import glob
import jieba

glob.glob('res_gossiping/*')
```

```python
sb = ''
for path in glob.glob('res_gossiping/*.txt'):
    with open(path, 'r')as f:
        sb += f.read().split("---split---")[0]

sb
```

```python
wcut = jieba.cut(sb)
```

```python
wcut = jieba.cut(sb)
wcut_list = [w for w in wcut]
wcut_list
```

```python
jieba.load_userdict('./mydict.txt')
```

```python
wordCountMap = dict()
for w in wcut_list:
    if w in wordCountMap:
        wordCountMap[w] += 1
    else:
        wordCountMap[w] = 1
wordCountMap
```

```python
wordCountList = [(k, v)for k, v in wordCountMap.items()]
wordCountList
```

```python
def sortLogic(x):
    return x[1]

wordCountList.sort(key=lambda x:x[1], reverse=True)
wordCountList
```
